```python
import requests
```

```python
res = requests.get('https://en.wikipedia.org/wiki/Grace_Hopper')
```

```python
import bs4
```

```python
soup = bs4.BeautifulSoup(res.text,'lxml')
```

```python
soup
```

```python
#toc -> table of content found while inspect wiki
```

```python
soup.select('.toctext')
```

```python
first_item = soup.select('.toctext')[0]
```

```python
#display all the text with span .toctext
for item in soup.select('.toctext'):
    print(item.text)
```
